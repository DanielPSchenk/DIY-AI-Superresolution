# Code for creating download lists
You should only need this code if you want to create your own dataset.

*NOTE THAT YOU CAN ONLY SEND 500 REQUESTS TO THE WIKIMEDIA COMMONS API PER HOUR*

```python


import requests

links = open("data/valued/link_list.txt").readlines()[400:899]
image_titles = ["File:" + link.split("File:")[1] for link in links]

output = open("data/valued/dl_links2.txt", "w")



for file in image_titles:
    url = 'https://api.wikimedia.org/core/v1/commons/file/' + file

    headers = {
        'User-Agent': 'YOUR WIKIMEDIA USER NAME'
    }

    response = requests.get(url, headers=headers)
    data = response.json()
    output.write(data["original"]["url"] + "\n")
    print(file)
output.close()

```

### Extracting links from webpage

```python
from bs4 import BeautifulSoup
file = open("data/valued/CVI.htm")
soup = BeautifulSoup(file.read())
```

```python
out_file = open("data/valued/link_list.txt", "w")
import validators

for link in soup.findAll('a'):
    url = link.get("href")
    if url != None and ("jpg" in url or "JPG" in url or "png" in url or "PNG" in url):
        if "wiki/File:" in url:
            out_file.write(url + "\n")
    
out_file.close()
```
